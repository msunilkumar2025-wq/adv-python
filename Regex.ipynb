```python
import regex as re
```

```python
pattern ="25"
text = "sunil 's birthday is on 25-11-2026"

match = re.search(pattern,text)
print(match)
print(type(match))
```

```python
pattern="627"
text="my registration no. is 24UG010627"

match=re.search(pattern,text)
print(match)
```

```python
string="hello 31 hi 56.Sunil 131"
pattern="\d+"

result=re.findall(pattern,string)
print(result)
```

```python
string="hello 31 hi 56.Sunil 131"
pattern="\d+"

result=re.finditer(pattern,string)
print(result)
```

```python
pattern ="Sunil"
text = "sunil 's birthday is on 25-11-2026"

match = re.search(pattern,text,re.IGNORECASE)
print(match)
print(type(match))
```
